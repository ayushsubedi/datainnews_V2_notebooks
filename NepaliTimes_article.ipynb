```python
import pandas as pd
import ast
import requests
from tqdm.auto import tqdm
from pandarallel import pandarallel
tqdm.pandas()
import nltk
# nltk.download('punkt')
from newspaper import Article
```

```python
pandarallel.initialize(progress_bar=True)
```

```python
df = pd.read_csv('NepaliTimes_keywords.csv')
```

```python
df
```

```python
def article(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return article.text
    except Exception as e:
        return 404
```

```python
df['article'] = df.urls.progress_apply(article)
```

```python
df.to_csv('NepaliTimes_articles.csv', index=False)
```
